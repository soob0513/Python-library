```python
# 데이터분석 라이브러리 로딩
import numpy as np    # 과학적인 수치 분석 전용 라이브러리
import pandas as pd   # 다수의 데이터를 표 형태로 저장하여 분석에 용이하게 하기 위한 라이브러리
```

```python
# 데이터 읽어오기! (조건 : 인덱스를 '관서명'으로 설정하고, 한글 인코딩('euc-kr'))
data19 = pd.read_csv('2019.csv', index_col = '관서명', encoding = 'euc-kr')
data20 = pd.read_csv('2020.csv', index_col = '관서명', encoding = 'euc-kr')
data21 = pd.read_csv('2021.csv', index_col = '관서명', encoding = 'euc-kr')
print(data19)
print(data20)
print(data21)
```

```python
# 2019,2020,2021 각각의 데이터 크기 확인 : shape
print(data19.shape)
print(data20.shape)
print(data21.shape)

# 2021년 데이터만 행이 조금 더 많다!
```

```python
# 특정 연도(2021년)에만 존재하는 경찰청인 '광주지방경찰청' 행 삭제 -> drop
data21.drop('광주지방경찰청',inplace = True)
```

```python
# 다시 한 번 크기 확인
print(data19.shape)
print(data20.shape)
print(data21.shape)
```

```python
# 1. 각 연도별 범죄데이터의 총합을 구해서 새로운 컬럼(열) 생성하기 ('총합')
# 단! 재실행 했을 때 총합 컬럼의 결과가 누적되지 않도록
data19['총합'] = data19.loc[:, '살인':'폭력'].sum(axis = 1)
data19
```

```python
data20['총합'] = data20.iloc[:, 1:6].sum(axis = 1)
data21['총합'] = data21.iloc[:, 1:6].sum(axis = 1)
```

```python
data20
```

```python
data21
```

```python
# 2. 구분 컬럼이 '발생건수'인 데이터들만 가져오기 -> 불리언 인덱싱
# temp2019 ~ temp2021 변수에 연도별 발생건수를 담아주기

temp2019 = data19[data19['구분'] == '발생건수']
temp2020 = data20[data20['구분'] == '발생건수']
temp2021 = data21[data21['구분'] == '발생건수']
temp2019
```

```python
temp2020
```

```python
temp2021
```

```python
# 3. 2번에 담아둔 데이터에서 '총합' 컬럼을 기준으로 데이터 가져오기(Series 형태 == 1차원 형태)
# s2019 ~ s2021 변수에 연도별로 각각 담아주기
s2019 = temp2019['총합']
s2020 = temp2020['총합']
s2021 = temp2021['총합']
s2021
```

```python
# 4. 3번에 담아둔 총합 데이터(Series)의 이름 바꿔주기
# '2019 총계', '2020 총계', '2021 총계'
s2019.name = '2019 총계'
s2020.name = '2020 총계'
s2021.name = '2021 총계'
s2019
```

```python
s2020
```

###  실습문제

### 결과 화면


```python
a20 = (s2020 - s2019) / s2019 * 100
a21 = (s2021 - s2020) / s2020 *100
a20.name = '2019-2020 증감율'
a21.name = '2020-2021 증감율'
```

```python
pd.concat([s2019,a20,s2020,a21,s2021],axis = 1)
```

```python
# 선생님풀이
# 범죄 증감율 계산
# (금년 - 작년) / 작년 * 100

# 2019 ~ 2020 증감율 구한 뒤 p_2020 변수에 담아주기
p_2020 = (s2020 - s2019) / s2019 * 100

# 2020 ~ 2021 증감율 구한 뒤 p_2021 변수에 담아주기
p_2021 = (s2021 - s2020) / s2020 * 100

p_2021
```

```python
# 증감율 데이터 이름 바꿔주기!
p_2020.name = '2019-2020 증감율'
p_2021.name = '2020-2021 증감율'
p_2021
```

```python
# 데이터프레임으로 병합하기 -> concat()

pd.concat([s2019,p_2020,s2020,p_2021,s2021], axis = 1)
```
