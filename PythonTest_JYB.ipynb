# 문제 4

```python
score_list = [['김미희', 89, 78, 99],
              ['서대희', 67, 88, 77],
              ['강예진', 100, 55, 96]]

평균 = 0
학점 =0
Python = 0
Python_1 = 0
Java = 0
Java_1 = 0
DB = 0
DB_1 = 0
띄어쓰기1 = 0
띄어쓰기2 = 0
띄어쓰기3 = 0
띄어쓰기4 = 0

print('=' * 49)
print('이름     python    Java     DB     평균     학점')
print('=' * 49)
for 이름 in score_list :
    평균 = (sum(이름[1:]))//(len(이름)-1)
    if 평균 >=90 :
        학점 = 'A'
    elif 평균 >= 80 :
        학점 = 'B'
    elif 평균 >= 70 :
        학점 = 'C'
    elif 평균 >= 60 :
        학점 = 'D'
    else :
        학점 = 'F'
    Python += 이름[1]
    Python_1 += 1
    Java += 이름[2]
    Java_1 += 1
    DB += 이름[3]
    DB_1 += 1

    if 이름[1] == 100 :
        띄어쓰기1 = ""
    else : 
        띄어쓰기1 = " "
    if 이름[2] == 100 :
        띄어쓰기2 = ""
    else : 
        띄어쓰기2 = " "
    if 이름[3] == 100 :
        띄어쓰기3 = ""
    else : 
        띄어쓰기3 = " "
    print(f'{이름[0]}    {이름[1]}{띄어쓰기1}       {이름[2]}{띄어쓰기2}     { 이름[3]}{띄어쓰기3}     {평균}{띄어쓰기3}      {학점}')
print('=' * 49)
print(f'평균      {Python//Python_1}        {Java//Java_1}      {DB//DB_1}')
```
