### 문제4

```python
score_list = [['김미희', 89, 78, 99],
             ['서대희', 67, 88, 77],
             ['강예진',100, 55, 96]]
score_list
```

```python
# 사람별 평균 구하는 함수
def h_avg(a) : 
    h_avg = sum(score_list[a][1:4]) / 3
    return round(h_avg,0)
```

```python
h_avg(1)
```

```python
# 학점 구하는 함수
def h_s(a) : 
    if h_avg(a) >= 90 : 
        s = 'A'
    elif h_avg(a) >= 80 : 
        s = 'B'
    elif h_avg(a) >= 70 : 
        s = 'C'
    elif h_avg(a) >= 60 : 
        s = 'D'
    else : 
        s = 'F'
    return s
```

```python
print('-'*80)
print('-'*80)
print('이름  Python  Java  DB   평균  학점')
print('-'*80)
print('-'*80)
for i in range(0,len(score_list)) : 
    print(score_list[i][0], score_list[i][1],score_list[i][2],score_list[i][3], h_avg(i), h_s(i))
print('-'*80)
print('-'*80)
```
